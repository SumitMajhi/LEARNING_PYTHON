```python
# Input validation
from math import sqrt
while True:
    user = int(input("Enter a number:"))
    if user < 0:
        print("Invalid input")
    elif user > 0:
        print(f"Square root of number is {sqrt(user)}")
    else:
        print("Ok, Bye!")
        break
print("Exiting program...")
```

```python
# Countdown
from time import sleep
for i in range(10,0,-1):
    print(i)
    sleep(1)
print("Blast off!")
```

```python
# Countdown
number = 5
print("Countdown!")
while True:
  print(number)
  number = number - 1
  if number <= 0:
    break

print("Now!")
```

```python
# Repeat password
attempts = 0
pin = input("Set your PIN: ")
while True:
    code = input("Please type in your PIN: ")
    attempts += 1

    if code == pin:
        success = True
        break

    if attempts == 3:
        success = False
        break

    # this is printed if the code was incorrect AND there have been less than three attempts
    print("Incorrect...try again")

if success:
    print("Correct PIN entered!")
else:
    print("Too many attempts...") 
```

```python
# The next leap year
year = int(input("year:"))
if (year% 4==0) or (year%4 == 0 and year% 100 != 0):
    print(f"The next leap year after {year} is {year+4}")
else:
    print("Not a leap year!")
```

```python
# The next leap year
u_year = int(input("Enter a year:"))
year = u_year + 1
while True:
    if (year% 4 == 0 and year% 100 != 0) or (year % 400 == 0):
        print(f"The next leap year after {u_year} is {year}")
        break
    year +=1
```

```python
# Word Story
sentence = []
while True:
    a = input("Enter a word:")
    if sentence and a == sentence[-1]:
        break
    elif a == "end":
        break
    sentence.append(a)
print(" ".join(sentence)) 
```
