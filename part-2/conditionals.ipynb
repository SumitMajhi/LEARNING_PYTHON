```python
# Typecasting
num = input("Please type in a number:")
if "." in num:
    num_int,dec_num = num.split(".")
else:
    num_int,dec_num = num,0
print(f"Integer part: {num_int}")
print(f"Decimal part: 0.{dec_num}")
```

```python
# Combining and chaining conditions
n1 = int(input("Number 1: "))
n2 = int(input("Number 2: "))
n3 = int(input("Number 3: "))
n4 = int(input("Number 4: "))

if n1 > n2 and n1 > n3 and n1 > n4:
    greatest = n1
elif n2 > n3 and n2 > n4:
    greatest = n2
elif n3 > n4:
    greatest = n3
else:
    greatest = n4

print(f" {greatest} is the greatest of the numbers.")
```

```python
# Grades and points
grade = int(input("How many points [0-100]: "))
if grade < 0:
    print("impossible!")
elif 0 <= grade <= 49:
    print("fail")
elif 50 <= grade <= 59:
    print("Grade: 1")
elif 60 <= grade <= 69:
    print("Grade: 2")
elif 70 <= grade <= 79:
    print("Grade: 3")
elif 80 <= grade <= 89:
    print("Grade: 4")
elif 90 <= grade <= 100:
    print("Grade: 5")
else:
    print("impossible!")
```

```python
# FizzBuzz
# If the number is divisible by three, the program should print out Fizz. 
# If the number is divisible by five, the program should print out Buzz. 
# If the number is divisible by both three and five, 
# the program should print out FizzBuzz.
num = int(input("Number:"))
if num% 3 == 0 and num% 5 == 0:
    print("FizzBuzz")
elif num% 3 == 0:
    print("Fizz")
elif num% 5 == 0:
    print("Buzz")
else:
    print("Not divisible")
```

```python
# Leap year
year = int(input("Please type in a year: "))
if (year% 400 ==0) or (year % 4 == 0 and year % 100 != 0):
    print("That year is a leap year.")
else:
    print("That year is not a leap year.")
```

```python
# Gift tax calculator
val = int(input("Value of gift:"))
if val >= 1000000 :
    print(f"Amount of tax: {142100 + ((val - 1000000)*0.17)} Rs")
elif 200000 <= val < 1000000:
    print(f"Amount of tax: {22100 + ((val - 200000)*0.15)} Rs")
elif 55000 <= val < 200000:
    print(f"Amount of tax: {4700 + ((val - 55000)*0.12)} Rs")
elif 25000 <= val < 55000:
    print(f"Amount of tax: {1700 + ((val - 25000)*0.10)} Rs")
elif 5000 <= val < 25000:
    print(f"Amount of tax: {100 + ((val - 5000)*0.08)} Rs")
else:
    print("No tax!")
```
