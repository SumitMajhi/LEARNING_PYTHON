```python
# The sum of consecutive numbers, version 1 (limit=10, 1+2+3+4= 10)
limit = int(input("Limit: "))
total = 0
number = 1
while total < limit:
    total += number
    number += 1
print(f"The total sum is {total}")
```

```python
# The sum of consecutive numbers, version 2
limit = int(input("limit: "))
total = 0
number = 1
expression = ""
while total < limit:
    total += number
    expression += str(number)

    if total < limit:
        expression += " + "
    number += 1
print(f"The sum of consecutive number {expression} is {total}")
    
```
