```python
# piramid
step = int(input("height of piramid: "))
row = "*"
while step > 0:
    print(" "* step + row)
    row += "**"
    step -= 1
```

```python
# string multiplied
text = input("Enter your word:")
times= int(input("Enter number of times to print:"))
print(text*times)
```

```python
# The longer string
text1 = input("Please type in string 1:")
text2 = input("Please type in string 2:")
if len(text1) > len(text2):
    print(f"{text1} is longer")
elif len(text1) < len(text2):
    print(f"{text2} is longer")
else:
    print("The strings are equally long")
```

```python
# print letters
text = input("Enter a string:")
index = 0
while index < len(text):
    print(text[index])
    index+=1
```

```python
# Reverse letters
text = input("Enter a string:")
index = len(text)-1
while index >= 0:
    print(text[index])
    index -=1
```

```python
# Second and second to last characters
text = input("Please type in a string:")
if text[1] == text[-2]:
    print(f"The second and the second to last characters are {text[1]}")
else:
    print("The second and the second to last characters are different")
```

```python
# A rectangle of hashes
w = int(input("Width:"))
h = int(input("Height:"))
while h-1 >= 0:
    print(w*"#")
    h -= 1
```

```python
# text with underline
while True:
    text = input("Enter a string to write and and press blank to exit:")
    print(text)
    print("-"*len(text))

    if text == "":
        break
```

```python
# Right-aligned
text = input("Enter a string:")
a = 20 - len(text)
if len(text) < 20:
    print("*"*a+text)
else:
    print(text)
```

```python
# A framed word
word = input("Word:")
width = 30
print("*"*width)

space = width - 2 - len(word)
left = space//2
right = space - left
print("*"+" "*left+word+" "*right+"*")

print("*"*width)
```
