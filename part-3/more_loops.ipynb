```python
num = int(input("Number : "))
while num > 0:
    i = 0
    while i < num:
        print(f"{i}", end="")
        i += 1
    print()
    num -=1
```

```python
# Multiplication
num = int(input("Number : "))
a =1
b =1
while a <= num:
    print(f"{a} x {b} = {a*b}")
    b +=1
    if b > num:
        a+=1
        b = 1
```

```python
# same code but simpler
num = int(input("Number:"))
for i in range(1,num+1):
    for j in range(1,num+1):
        print(f"{i} x {j} = {i*j}")
```

```python
# Factorial
while True:
    num = int(input("Please type in a number: "))

    if num <= 0:
        print("Thanks and bye!")
        break

    fact = 1
    for i in range(1, num + 1):
        fact *= i

    print(f"The factorial of the number {num} is {fact}")
```

```python
while True:
    num = int(input("Enter a number:"))
    if num <= 0:
        print("Thanks and Bye")
        break
    fact = 1
    for i in range(1,num+1):
        fact *= i
    print(f"Factorial of {num} is {fact}")
```

```python
# Flip the pairs
num = int(input("Enter a number:"))
for i in range(1,num+1,2):
    if i+1 <= num:
        print(i+1)
        print(i)
    else:
        print(i)
```

```python
num = int(input("Enter a number:"))
left = 1
right = num
while left <= right:
    print(left)

    if left != right:
        print(right)

    left += 1
    right -=1
```
