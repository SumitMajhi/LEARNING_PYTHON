```python
# seven brothers
def seven_brothers():
    names = ['Aapo','Eero','Juhani','Lauri','Simeoni','Timo','Tuomas']
    names.sort()
    for name in names:
        print(name)

if __name__ == "__main__":
    seven_brothers()
```

```python
# squared hash
def squared_hash(a):
    for _ in range(a):
        print("#"*a)
if __name__=="__main__":
    squared_hash(5)
```

```python
# chessboard
def chessboard(size):
    for row in range(size):
        line =""
        for col in range(size):
            if (row+col)% 2 == 0:
                line+="1"
            else:
                line+="0"
        print(line)

chessboard(5)
```

```python
# squared
def squared(text,size):
    index = 0
    for row in range(size):
        line = ""
        for col in range(size):
            line+=text[index]
            index+=1
        
            if index == len(text):
                index = 0
        print(line)

squared("alice",3)
```
