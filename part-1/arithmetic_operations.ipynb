```python
# Food expenditure
eat = int(input("How many times a week do you eat at the student cafeteria?"))
price = float(input("The price of a typical student lunch?"))
s_weak = float(input("How much money do you spend on groceries in a week?"))
print(f"Average food expenditure:\nDaily: {((eat*price)+s_weak)/7} euros \nWeekly: {(eat*price)+s_weak} euros")
```

```python
# Students in groups
stu = int(input("How many students on the course?"))
size = int(input("Desired group size?"))
grp = 0
if (stu/size)%2 == 0:
    grp = (stu//size)
else:
    grp = (stu//size)+1
print(f"Number of groups formed: {grp}")
```
