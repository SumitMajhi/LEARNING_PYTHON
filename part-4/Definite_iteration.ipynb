```python
#Anagrams
def anagrams(a,b):
    return sorted(a) == sorted(b)
print(f"{anagrams("ansi","sani")}")
```

```python
# Palindromes
def palindromes(text : str):
    return text == text[::-1]

while True:
    text = input("Please type in a palindrome:")
    if palindromes(text):
        print(f"{text} is a palindrome!")
        break
    else:
        print("that wasn't a palindrome")
        
```

```python
#The sum of positive numbers
def sum_of_positive(my_list:list):
    i = 0
    for num in my_list:
        if num > 0:
            i+= num
    return i
my_list= [1,-4,6,-2,7,3]
result = sum_of_positive(my_list)
print(result)
```

```python
# Even numbers
def even_numbers(my_list:list):
    new_list = []
    for num in my_list:
        if num % 2 ==  0:
            new_list.append(num)
    return new_list
my_list = [1,2,3,4,5,6]
print(my_list)
result = even_numbers(my_list)
print(result)
```

```python
# The sum of lists
def list_sum(a:list, b:list):
    result = []
    for i in range (len(a)):
        result.append(a[i]+b[i])
    return result

a = [1,2,3,4,5]
b = [1,2,3,4,5]
out = list_sum(a, b)
print(out)
```

```python
# distinct_numbers(my_list)
def distinct_numbers(my_list:list):
    result = []
    for item in my_list:
        if item not in result:
            result.append(item)
    result.sort()
    return result
if __name__ == "__main__":
    my_list = [2,4,2,4,3,3,6,3,45,6]
    print(distinct_numbers(my_list))
```

```python
#The length of the longest in the list
def length_of_longest(my_list:list):
    longest = 0
    for char in my_list:
        if len(char) > longest:
            longest = len(char)
    return longest

my_list = ["first", "second", "fourth", "eleventh"]
result = length_of_longest(my_list)
print(result)
```

```python
# The shortest in the list
def shortest(my_list:list):
    short_word = my_list[0]
    for word in my_list:
        if len(word) < len(short_word):
            short_word = word
    return short_word

my_list = ["first", "second", "fourth", "eleventh"]

result = shortest(my_list)
print(result)
```

```python
# All the longest in the list
def  all_the_longest(my_list:list):
    longest = 0
    new = []
    for word in my_list:
        if len(word) > longest:
            longest = len(word)

    for char in my_list:
        if len(char) == longest:
            new.append(char)
    
    return new

my_list = ["adele", "mark", "dorothy", "tim", "hedy", "richard"]

result = all_the_longest(my_list)
print(result)
```
