```python
# Everything reversed
def everything_reversed(my_list:list):
    new = []
    for word in my_list:
        word_rev = word[::-1]
        new.append(word_rev)
    
    return new[::-1]
my_list = ["Hi", "there", "example", "one more"]
new_list = everything_reversed(my_list)
print(new_list)
```

```python
#Most common character
def most_common_character(my_str:str):
    most_common = my_str[0]
    for character in my_str:
        if my_str.count(character) > my_str.count(most_common):
            most_common = character
    return most_common
second_string = "exemplaryelementary"
print(most_common_character(second_string))
```

```python
#No vowels allowed
def no_vowels(my_str:str):
    vowels = "aeiou"
    new_str = ""
    for character in my_str:
        if character not in vowels:
            new_str += character
    return new_str

my_string = "this is an example"
print(no_vowels(my_string))
```

```python
# No shouting
def no_shouting(my_list:list):
    new = []
    for word in my_list:
        if word.isupper() == False:
            new.append(word)
    return new

my_list = ['FIRST', 'second', 'THIRD', 'fourth']
print(no_shouting(my_list))
```

```python
# neighbours in the list
def neighbours(my_list:list):
    current = 1
    longest = 1
    for i in range (1,len(my_list)):
        if abs(my_list[i] - my_list[i-1]) == 1:
            current +=1
        else:
            current = 1

        if current > longest:
            longest = current
    return longest

my_list = [1, 2, 5, 7, 6, 5, 6, 3, 4, 1, 0]
print(neighbours(my_list))
```

```python
def input_from_user(how_many: int):
    print(f"Please type in {how_many} numbers:")
    numbers = []

    for i in range(how_many):
        number = int(input(f"Number {i+1}: "))
        numbers.append(number)

    return numbers

def print_result(numbers: list):
    print("The numbers are: ")
    for number in numbers:
        print(number)

def analyze(numbers: list):
    mean = sum(numbers) / len(numbers)
    return f"There are altogether {len(numbers)} numbers, the mean is {mean}, the smallest is {min(numbers)} and the greatest is {max(numbers)}"

# the "main function" using these functions
def main():
    inputs = input_from_user(5)
    print_result(inputs)
    analysis_result = analyze(inputs)

    print(analysis_result)

# run the main function
main()
```

```python
def exercise_to_points(exercises: int):
    # 10% = 1 point, 100% = 10 points (integer division)
    return exercises // 10

def determine_grade(exam_points: int, exercises: int):
    exer_points = exercise_to_points(exercises)
    total = exam_points + exer_points
    
    # Exam cutoff: less than 10 points is an automatic fail (grade 0)
    if exam_points < 10:
        return 0
    
    if total <= 14:
        return 0
    elif total <= 17:
        return 1
    elif total <= 20:
        return 2
    elif total <= 23:
        return 3
    elif total <= 27:
        return 4
    else:
        return 5

def main():
    all_totals = []
    grades = [0] * 6 # To store counts for grades 0-5
    
    while True:
        entry = input("Exam points and exercises completed: ")
        if not entry:
            break
            
        parts = entry.split()
        exam = int(parts[0])
        exer = int(parts[1])
        
        # Calculate totals for the average
        all_totals.append(exam + exercise_to_points(exer))
        
        # Determine grade and update distribution
        grade = determine_grade(exam, exer)
        grades[grade] += 1

    print("Statistics:")
    
    # 1. Points Average
    avg = sum(all_totals) / len(all_totals)
    print(f"Points average: {avg:.1f}")
    
    # 2. Pass Percentage
    # Grade 0 is a fail; grades 1-5 are passes
    passed = sum(grades[1:])
    pass_pct = (passed / len(all_totals)) * 100
    print(f"Pass percentage: {pass_pct:.1f}")
    
    # 3. Grade Distribution
    print("Grade distribution:")
    for i in range(5, -1, -1):
        stars = "*" * grades[i]
        print(f"  {i}: {stars}")

# Run the program
main()
```
