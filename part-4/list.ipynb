```python
# Change the value of an item
my_list=[1,2,3,4,5]
while True:
    index = int(input("Index :"))
    if index == -1:
        break

    new_value = int(input("New Value: "))
    my_list[index] = new_value
    print(my_list)
```

```python
# Add items to a list
add_item = int(input("How many items:"))
my_list = []
for i in range(add_item):
    item = int(input(f"Item {i+1}"))
    my_list.append(item)
print(my_list)
```

```python
# Addition and removal
items = []
while True:
    print(f"The list is now {items}")
    choice = input("a(d)d, (r)emove, e(x)it :")

    if choice == "d":
        if len(items) == 0:
            items.append(1)
        else:
            items.append(items[-1] + 1)
    elif choice == "r":
        if len(items) > 0:
            items.pop(-1)

    elif choice == "x":
        break

print("Bye!")
```

```python
#Same word twice
words = []
count = 0
while True:
    word = input("Word:")
    if word in words:
        break
    
    words.append(word)
    count +=1
    
print(f"You typed {count} different words")
```

```python
#List twice
now = []
order = []
while True:
    item = input("New Item:")
    if item == "0":
        break
    now.append(item)
    order.append(item)
    order.sort()
    print(f"The list now:{now}")
    print(f"The list in order:{order}")
print("Bye!")
```

```python
numbers = []
while True:
    item = int(input("Item:"))
    if item == 0:
        break
    numbers.append(item)
    print(f"The list now :{numbers}")
    print(f"The order list:{sorted(numbers)}")
print("Bye!")
```
