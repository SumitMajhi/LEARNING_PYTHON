```python
# print pattern
def line(pattern,times):
    if pattern == "":
        pattern = "*"
    print(pattern[0]*times)

line("^",6)
line("!",8)
```

```python
# box of hashes
def line(pattern,times):
    print(pattern*times)

def hash_box(rep):
    while rep > 0:
        line("#",10)
        rep-=1

hash_box(5)
```

```python
#squared hash
def line(pattern,times):
    print(pattern*times)

def square_of_hashes(size):
    for i in range(size):
        line("#",size)

if __name__ == "__main__":
    square_of_hashes(5)
```

```python
def line(pattern,times):
    print(pattern*times)

def squared(character,size):
    for i in range(size):
        line(character,size)

squared("<>",5)
```

```python
# Triangle
def line(pattern,times):
    print(pattern*times)

def tri(size):
    for i in range(1,size+1):
        line("^",i)

tri(5)
```

```python
# shape
def line(char,size):
    print(char*size)

def triangle(char,size):
    for i in range(1,size+1):
        line(char,i)

def shape(size1,char1,size2,char2):
    triangle(char1,size1)

    for _ in range(size2):
        line(char2,size1)

shape(4,"$",6,"x")
```
