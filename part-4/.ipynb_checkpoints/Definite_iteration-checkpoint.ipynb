```python
#Star-studded
a =  "Sumit"  #input("Enter text:")
index = 0
while True:
    if index < len(a):
        print(f"{a[index]} \n*")
        index +=1
    else:
        break
```

```python
a = "Input"
for char in a:
    print(f"{char} \n*")
```

```python
#From negative to positive
a = int(input("Enter positive number:"))
for i in range(-a,a+1):
    if i != 0:
        print(i)
```

```python
def list_of_stars(my_list: list):
    for number in my_list:
        print("*" * number)
my_list = [3,6,7,2,1]
list_of_stars(my_list)
```

```python
#Anagrams
def anagrams(a,b):
    return sorted(a) == sorted(b)
print(f"{anagrams("ansi","sani")}")
```

```python
# Palindromes
def palindromes(text : str):
    return text == text[::-1]

while True:
    text = input("Please type in a palindrome:")
    if palindromes(text):
        print(f"{text} is a palindrome!")
        break
    else:
        print("that wasn't a palindrome")
        
```
